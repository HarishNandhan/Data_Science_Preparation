```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1
```

```python
a = pd.get_dummies(df1['RestingECG'])
a
```

```python
df1 = pd.concat([df1,a],axis=1)
df1
```

```python
df1.drop('RestingECG',axis=1, inplace=True)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder

le_chesttype = LabelEncoder()
le_sex = LabelEncoder()
le_exer = LabelEncoder()
le_slp = LabelEncoder()
```

```python
df1['Sex'] = le_sex.fit_transform(df1['Sex'])
df1['ChestPainType'] = le_chesttype.fit_transform(df1['ChestPainType'])
df1['ExerciseAngina'] = le_exer.fit_transform(df1['ExerciseAngina'])
df1['ST_Slope'] = le_slp.fit_transform(df1['ST_Slope'])
```

```python
df1
```

```python
X = df1.drop('HeartDisease',axis=1)
y = df1.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

scaler  = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X_scaled, y)
```

```python
x = pd.DataFrame(X_pca) 
x

# From 15 columns it's being reduced to 11 columns while retaining 95% of the information (95% variance)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores_svm = cross_val_score(SVC(), X_scaled, y, cv=5)
scores_svm.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores_dt = cross_val_score(DecisionTreeClassifier(), X_scaled, y, cv=5)
scores_dt.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model_svm = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=42
)

bag_model_dt =  BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=42
)

bag_model_svm.fit(X_scaled, y)
bag_model_svm.oob_score_
```

```python
bag_model_dt.fit(X_scaled, y)
bag_model_dt.oob_score_
```
