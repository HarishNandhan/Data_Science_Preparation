```python
import pandas as pd

df = pd.read_csv("homeprices.csv")
df.head(10)
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns')
merged
```

```python
merged.drop(columns=['town'],axis = 1)
```

```python
merged.head(10)
```

Out of the Dummy columns generated we have to mandatorily drop either one of the dummy variable column since it may create the multicollinearity in the dataframe and that can lead to the bias while developing a Machine Learning Model.


```python
final = merged.drop(columns=['town','west windsor'],axis = 1)
final
```

```python
final.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price',axis=1)
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]]) # In Robinsville
```

```python
model.predict([[3400,0,0]]) # For west windsor
```

```python
model.score(X,y) # Accuracy of the Model
```
