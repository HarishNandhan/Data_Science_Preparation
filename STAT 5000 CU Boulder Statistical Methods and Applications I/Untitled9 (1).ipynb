```python
import pandas as pd 
```

```python
data = pd.read_csv("fraud_credit_card.csv")
data
```

```python
columns_to_drop = ['Unnamed: 0', 'cc_num', 'first', 'last', 'street', 'city', 'state','trans_num', 'zip', 'job','unix_time', 'dob']
data_cleaned = data.drop(columns=columns_to_drop)

data_cleaned['trans_date_trans_time'] = pd.to_datetime(data_cleaned['trans_date_trans_time'], format='%d/%m/%Y %H:%M')
data_cleaned['year'] = data_cleaned['trans_date_trans_time'].dt.year
data_cleaned['month'] = data_cleaned['trans_date_trans_time'].dt.month
data_cleaned['day'] = data_cleaned['trans_date_trans_time'].dt.day
data_cleaned['hour'] = data_cleaned['trans_date_trans_time'].dt.hour
data_cleaned = data_cleaned.drop(columns=['trans_date_trans_time'])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

categorical_columns = ['merchant', 'category', 'gender']
for col in categorical_columns:
    data_cleaned[col] = label_encoder.fit_transform(data_cleaned[col])

data_cleaned.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_columns = ['amt', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
data[numerical_columns]
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE 
from sklearn.model_selection import train_test_split

# Split features and target variable
X = data_cleaned.drop(columns=['is_fraud'])
y = data_cleaned['is_fraud']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

```python
# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Initialize ensemble models
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
adaboost = AdaBoostClassifier(n_estimators=100, random_state=42)

# Train the Random Forest model (Bagging)
random_forest.fit(X_train, y_train)
y_pred_rf = random_forest.predict(X_test)

# Train the AdaBoost model (Boosting)
adaboost.fit(X_train, y_train)
y_pred_ab = adaboost.predict(X_test)

# Evaluate Random Forest
confusion_rf = confusion_matrix(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# Evaluate AdaBoost
confusion_ab = confusion_matrix(y_test, y_pred_ab)
precision_ab = precision_score(y_test, y_pred_ab)
recall_ab = recall_score(y_test, y_pred_ab)
f1_ab = f1_score(y_test, y_pred_ab)

# Print evaluation metrics for Random Forest
print("Random Forest (Bagging) Confusion Matrix:\n", confusion_rf)
print("Random Forest Precision:", precision_rf)
print("Random Forest Recall:", recall_rf)
print("Random Forest F1 Score:", f1_rf)

# Print evaluation metrics for AdaBoost
print("\nAdaBoost (Boosting) Confusion Matrix:\n", confusion_ab)
print("AdaBoost Precision:", precision_ab)
print("AdaBoost Recall:", recall_ab)
print("AdaBoost F1 Score:", f1_ab)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns

# Define the hyperparameter grid
param_grid_dt = {
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5, 10]
}

# Initialize the Decision Tree model
decision_tree = DecisionTreeClassifier(random_state=42)

# Perform Grid Search
grid_search_dt = GridSearchCV(decision_tree, param_grid_dt, cv=5, scoring='f1', n_jobs=-1)
grid_search_dt.fit(X_train, y_train)

# Best model from Grid Search
best_dt = grid_search_dt.best_estimator_

# Predict on the test set
y_pred_dt = best_dt.predict(X_test)

# Generate and plot the confusion matrix
confusion_dt = confusion_matrix(y_test, y_pred_dt)
ConfusionMatrixDisplay(confusion_matrix=confusion_dt).plot(cmap='Blues')
plt.title("Decision Tree Confusion Matrix")
plt.show()

# Best hyperparameters
print("Best Hyperparameters for Decision Tree:", grid_search_dt.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression

# Define the hyperparameter grid
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear']
}

# Initialize the Logistic Regression model
logistic_regression = LogisticRegression(max_iter=1000, random_state=42)

# Perform Grid Search
grid_search_lr = GridSearchCV(logistic_regression, param_grid_lr, cv=5, scoring='f1', n_jobs=-1)
grid_search_lr.fit(X_train, y_train)

# Best model from Grid Search
best_lr = grid_search_lr.best_estimator_

# Predict on the test set
y_pred_lr = best_lr.predict(X_test)

# Generate and plot the confusion matrix
confusion_lr = confusion_matrix(y_test, y_pred_lr)
ConfusionMatrixDisplay(confusion_matrix=confusion_lr).plot(cmap='Blues')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

# Best hyperparameters
print("Best Hyperparameters for Logistic Regression:", grid_search_lr.best_params_)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Define the hyperparameter grid
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

# Initialize the KNN model
knn = KNeighborsClassifier()

# Perform Grid Search
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=5, scoring='f1', n_jobs=-1)
grid_search_knn.fit(X_train, y_train)

# Best model from Grid Search
best_knn = grid_search_knn.best_estimator_

# Predict on the test set
y_pred_knn = best_knn.predict(X_test)

# Generate and plot the confusion matrix
confusion_knn = confusion_matrix(y_test, y_pred_knn)
ConfusionMatrixDisplay(confusion_matrix=confusion_knn).plot(cmap='Blues')
plt.title("KNN Confusion Matrix")
plt.show()

# Best hyperparameters
print("Best Hyperparameters for KNN:", grid_search_knn.best_params_)
```

```python
from sklearn.svm import SVC

# Define the hyperparameter grid
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

# Initialize the SVM model
svm = SVC(random_state=42)

# Perform Grid Search
grid_search_svm = GridSearchCV(svm, param_grid_svm, cv=5, scoring='f1', n_jobs=-1)
grid_search_svm.fit(X_train, y_train)

# Best model from Grid Search
best_svm = grid_search_svm.best_estimator_

# Predict on the test set
y_pred_svm = best_svm.predict(X_test)

# Generate and plot the confusion matrix
confusion_svm = confusion_matrix(y_test, y_pred_svm)
ConfusionMatrixDisplay(confusion_matrix=confusion_svm).plot(cmap='Blues')
plt.title("SVM Confusion Matrix")
plt.show()

# Best hyperparameters
print("Best Hyperparameters for SVM:", grid_search_svm.best_params_)
```
