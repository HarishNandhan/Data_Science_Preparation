### Exercise 1: Histograms of Maximum Temperature 

Suppose we want to draw a histogram of the **TMAX** characteristics over all weather stations.  There are several ways to plot histograms in R. We will investigate base R methods in this exercise and ggplot methods in the next exercise.

```python
# plot histogram in base R
# 1) Code Here
library(ggplot2)
hist(weather_data$TMAX,
    xlab = "Temperature in Fahrenheit",
    xlim = c(30,110),
     col= "red",
     border = 13.5,
    freq = FALSE)
```

```python
# histogram in base R with added parameters
# 2) Code Here
library(ggplot2)
h = hist(weather_data$TMAX,
    main = "Max Daily Temp in Boulder",
    xlab = "Temperature in Fahrenheit",
    xlim = c(30,110),
     col= "blue",
     border = 13.5,
    freq = NULL)

text(h$mids, h$counts, labels=h$counts, adj=c(0.5,-0.5))
```

### Exercise 2: Histograms of Maximum Temperature - ggplot

Let's explore how to create and customize histograms using ggplot. First, we import the `ggplot2` library.

```python
library(ggplot2)
```

```python
#basic histogram
# 4) Code Here
ggplot(weather_data, aes(x=TMAX)) + geom_histogram()
```

```python
# change bin width and add colors
# 5) Code Here

plot = ggplot(weather_data, aes(x=TMAX)) + 
        geom_histogram(bindwidth = 1, color = "Black",fill="darkmagenta",show.legend = TRUE)
plot
```

Next, let's add a mean line and a density plot overlay. 

```python
# mean line
# 6) Code Here
plot + geom_vline(aes(xintercept = mean(TMAX)), color="red",linetype="dashed",
                 linewidth = 50)
```

```python
# density plot overlay
# 7) Code Here
ggplot(weather_data, aes(x=TMAX))+
    geom_histogram(aes(y=..density..),color="black",fill="white", bandwidth=1)+
    geom_density(aplha=.4, fill="blue")
```
