```python
df['Status'] = np.where(df['Status'] == 'Alive', 0, 1)
```

Here, this code creates a boolean mask for the 'Status' column and it will assign '0' for Alive Cases and it will assign '1' for the Dead Cases.

##### One Hot Encoding

```python
def one_hot_encoding(df, categorical_features):
    unique_values = {col: df[col].unique().tolist() for col in categorical_features}
    
    encoded_data = [
        [1 if row[col] == unique else 0 for col in categorical_features for unique in unique_values[col]]
        for _, row in df.iterrows()
    ]
    
    return np.array(encoded_data)

encoded_categorical = one_hot_encoding(df, categorical_features)
```

- Here we are extracting the unique categories present in the data by creating a dictionary 'unique_values' where each category column is mapped with its unique values.
- Looping through every row in the dataframe 'df' to generate a one hot encoded representation.
- Creating the one hot encoding vectors for each and every categorical features, checking if the value matches a unique category assigning '1' (match) or 0 (not a match).
- Return the numpy array which stored the one hot encoded data,


##### Standardization

```python
def standardize(data):
    return (data - np.mean(data, axis=0)) / np.std(data, axis=0)

scaled_continuous = standardize(df[continuous_features].to_numpy())
```

- Here in this function, the data is centered by subtracting the mean and we are scaling it by dividing by the Standard deviation for each and every feature which ensures all the continuous features have zero mean and unit variance (Standard Normal Distribution).
- This function is applied to the df['continuous_features'] which coverts the continuous columns to numpy arrays before the standardization process making it ready for the distance calculation.

##### Splitting the dataset into training, testing and validation

```python
X = np.hstack((scaled_continuous, encoded_categorical))
y = df['Status'].values

def train_test_split(X, y, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15, seed=32):
    np.random.seed(seed) 
    indices = np.random.permutation(len(X)) 


    train_end = int(len(X) * train_ratio)
    val_end = train_end + int(len(X) * val_ratio)


    return (
        X[indices[:train_end]], y[indices[:train_end]],  
        X[indices[train_end:val_end]], y[indices[train_end:val_end]], 
        X[indices[val_end:]], y[indices[val_end:]] 
    )

X_train, y_train, X_val, y_val, X_test, y_test = train_test_split(X, y)
```

- Here Combining the scaled continuous features and the one hot encoded categorical features into a single feature matrix (x).
- Extracting the target variable (y=df['Status'].values), where the status = Alive/Dead is the classification label.
- The np.random.permutation(len(X)) which randomly shuffles the indices, so that the data points will be randomly shuffled before the splitting.
- Splitting the dataset into 70% training, 15% validation and 15% testing based on the computed indices of 'train_end' and 'val_end'
- At the end, X_train, y_train will be used for the model training, X_val, y_val will be used for the hyper parameter tuning/validation and the X_test, y_test will be used for the Model evaluation.

##### Computing the Euclidean distance for Continuous variables & Hamming Distance for Categorical Variables

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def hamming_distance(x1, x2):
    return np.sum(x1 != x2)
```

- Here the function 'euclidean_distance(x1,x2)' computes the euclidean distance between two numerical feature point vectors by taking the square root of the sum of squared differences.
- Here the function 'hamming_distance(x1, x2)' counts the no. of positions where the two categorical feature vectors vary.

##### Computing the KNN Predicting Function

```python
def knn_predict(X_train, y_train, X_test, K):
    y_pred = []  

    for test_sample in X_test:  
        distances = [
            (
                euclidean_distance(train_sample[:4], test_sample[:4]) + 
                hamming_distance(train_sample[4:], test_sample[4:]),  
                train_label
            )
            for train_sample, train_label in zip(X_train, y_train)  
        ]

        
        k_nearest_labels = [label for _, label in sorted(distances, key=lambda x: x[0])[:K]]

        
        y_pred.append(max(set(k_nearest_labels), key=k_nearest_labels.count))

    return np.array(y_pred)
```

- Here the 'knn_predict' function computes the distance for each and every test sample, it calculates the sum of Euclidean distance and Hamming Distance from all the training samples.
- It finds the K nearest neightbors by sorting the training samples by distance and it will select the K closest neighbor labels.
- Doing majority voting by assigning the most frequent label among the K neighbors as the predicted class for the test sample.

##### Computing the Evaluation Metrics & Hyper Parameter Tuning to find out the value for best value of K based on Validation Set

```python
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred) 


def balanced_accuracy(y_true, y_pred):
    unique_classes = np.unique(y_true)
    
    class_accuracies = [
        np.mean(y_pred[y_true == label] == label) 
        for label in unique_classes
    ]
    
    return np.mean(class_accuracies)


def f1_score(y_true, y_pred, positive_class=1):
    tp = np.sum((y_true == positive_class) & (y_pred == positive_class))
    fp = np.sum((y_true != positive_class) & (y_pred == positive_class))
    fn = np.sum((y_true == positive_class) & (y_pred != positive_class))
    
    precision = tp / (tp + fp) if (tp + fp) else 0
    recall = tp / (tp + fn) if (tp + fn) else 0
    
    return (2 * precision * recall) / (precision + recall) if (precision + recall) else 0


K_values = [1, 3, 5, 7, 9, 11, 13]
results = [
    {
        'K': K,
        'Accuracy': accuracy(y_val, knn_predict(X_train, y_train, X_val, K)),
        'Balanced Accuracy': balanced_accuracy(y_val, knn_predict(X_train, y_train, X_val, K)),
        'F1-Score': f1_score(y_val, knn_predict(X_train, y_train, X_val, K))
    }
    for K in K_values
]


res_df = pd.DataFrame(results)
print(res_df)
```

- Defining a function for accuracy, balanced_accuracy and f1_score.
- Looping throught the k values here i took k = [1,3,5,7,9,11,13]
- This function calls the 'knn_predict' function using X_train, X_val to generate the predictions of the validation set at each K value computation
- The resulting accuracy, balanced accuracy and f1_Score for each value of k is stored in a dataframe.
- The best k value
  - based on 'Accuracy' is k = 13
  - based on 'Balanced Accuracy' is k = 13
  - based on 'F1_score' is k = 5

```python
K_values = res_df["K"]
accuracy = res_df["Accuracy"]
balanced_accuracy = res_df["Balanced Accuracy"]
f1_score_values = res_df["F1-Score"]

plt.figure(figsize=(6, 4))
plt.plot(K_values, accuracy, marker='o', linestyle='-', label="Accuracy")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. K")
plt.xticks(K_values)
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(K_values, balanced_accuracy, marker='o', linestyle='-', label="Balanced Accuracy", color='orange')
plt.xlabel("K")
plt.ylabel("Balanced Accuracy")
plt.title("Balanced Accuracy vs. K")
plt.xticks(K_values)
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(K_values, f1_score_values, marker='o', linestyle='-', label="F1-Score", color='green')
plt.xlabel("K")
plt.ylabel("F1-Score")
plt.title("F1-Score vs. K")
plt.xticks(K_values)
plt.grid(True)
plt.show()
```

- For plotting, retrieving the 'k_values', 'accuracy', 'balanced_accuracy' and 'f1_score' from the resulting dataframe.
- These plots are helpful in visualizing how the model performance changes with the different values of K.

#### f) MODEL DEPLOYMENT

```python
best_k_f1 = 5  
df_test = df.iloc[np.random.permutation(len(df))[-len(y_test):]].reset_index(drop=True)

def filter_test_by_race(X_test, y_test, df_test, race_value):
    """Filters X_test and y_test based on race from the aligned test dataset"""
    race_mask = df_test["Race"] == race_value
    if race_mask.sum() == 0:
        return None, None
    return X_test[race_mask], y_test[race_mask]

race_f1_scores = {}

for race in ["White", "Black", "Other"]:
    X_race, y_race = filter_test_by_race(X_test, y_test, df_test, race)

    if X_race is not None and len(y_race) > 0:
        y_race_pred = knn_predict(X_train, y_train, X_race, best_k_f1)


        race_f1_scores[race] = f1_score(y_race, y_race_pred)
    else:
        race_f1_scores[race] = None

race_f1_df = pd.DataFrame(list(race_f1_scores.items()), columns=["Race", "F1-Score"])

print(race_f1_df)
```

##### Effect of Class imbalance on Evaluation Metrics:
- Class imbalance affects the accuracy of the model because here the majority case may make the True Positive class more and it will have higher accuracy but in reality the performance of predicting the poor class will be extremely low.
- Balanced class accuracy is best for evaluating the Binary Classification with imbalanced dataset but even here if the minority class is extremely low then this metric also won't work accurately.
- The class imbalance also affects the F1 score because F1 score depends on precision and recall and if the minority class samples are extremely low then the recall will also be low which leads to a lower F1 score.

##### Methods of Handling Class Imbalance Issues:
- Upsampling techniques like Synthetic Minority Oversampling Technique (SMOTE) is useful in generating duplicate the original or generating new synthetic samples.
- Downsampling helps in reducing the random samples from the majority class to make sure equal balance exists between both of the classes.
- Data Augumentation can be used for balancing the datasets in which the data is of the form of images.
- Class weighted models are useful in assigning higher weights to minority classes which inturn reduces the bias towards the majority class samples.
